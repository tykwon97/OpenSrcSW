```python
#실습 1 : 생일 기온 시각화

import csv
import matplotlib.pyplot as plt

f = open('week13_data/seoul.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)
result_max = []
result_min = []
for row in data :
    
    if (row[-1] != '' and row[-2] != '') : #최고 기온, 최저 기온
        if (row[0].split('-')[1] == '12' and row[0].split('-')[2] == '19' ) :
            result_max.append(float(row[-1]))
            result_min.append(float(row[-2]))
f.close()
plt.plot(result_max)
plt.plot(result_min)

plt.show()
```

```python
#실습 2 : 최고 기온 시각화
import csv
import matplotlib.pyplot as plt

f = open('week13_data/seoul.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)
result_jan = []
result_aug = []
 
for row in data :
    if (row[-1] != '') : #최고 기온
        if (row[0].split('-')[1] == '01') :
            result_jan.append(float(row[-1]))
        if (row[0].split('-')[1] == '08') :
            result_aug.append(float(row[-1]))
f.close()

plt.title("December 19th")
plt.hist(result_jan,bins=100,color="blue",label="Jan")
plt.hist(result_aug,bins=100,color="red",label="Aug")
plt.legend() #범례 표시
plt.show()
```
