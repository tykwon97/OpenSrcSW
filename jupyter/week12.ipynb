```python
#실습 1 : 서울의 일교차
# 기온 공공 데이터를 이용하여 서울에서 일교차가 가장 큰 날, 최저기온이 가장 낮은 날, 최고 기온이 가장 높을 날을 찾아 출력하시오.
import csv

f = open('week12_data/seoul.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)    
difference_temp = -999
difference_date = ''
for row in data:
    if(row[-1] == ''):
        row[-1] = 0
    if(row[-2] == ''):
        row[-2] = 0
    row[-1] = float(row[-1])
    row[-2] = float(row[-2])
    if(difference_temp < row[-1]-row[-2]):
        difference_date = row[0]
        difference_temp = row[-1]-row[-2]
f.close()      

#최고 기온
f = open('week12_data/seoul.csv','r',encoding='cp949')
data = csv.reader(f)#,delimiter=',')
header = next(data)   
max_temp = -999
max_date = ''
for row in data:
    if(row[-1] == ''):
        row[-1] = -999
    row[-1] = float(row[-1])
    if(max_temp < row[-1]):
        max_date = row[0]
        max_temp = row[-1]
f.close()        
        
#최저 기온 
f = open('week12_data/seoul.csv','r',encoding='cp949')
data = csv.reader(f)#,delimiter=',')
header = next(data)   
min_temp = 999
min_date = ''
for row in data:
    if(row[-2] == ''):
        row[-2] = 999
    row[-2] = float(row[-2])
    if(min_temp > row[-2]):
        min_date = row[0]
        min_temp = row[-2]
f.close()

print('기상 관측 이래 서울의 일교차가    가장 큰 날은     {0} 로, {1} 도 였습니다.'.format(difference_date,difference_temp))
print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은 {0} 로, {1} 도 였습니다.'.format(max_date,max_temp))
print('기상 관측 이래 서울의 최저 기온이 가장 낮았던 날은 {0} 로, {1} 도 였습니다.'.format(min_date,min_temp))
```
